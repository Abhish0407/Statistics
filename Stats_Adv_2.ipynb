1] What is hypothesis testing in statistics
- Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It involves forming a null hypothesis (usually stating no effect or difference) and an alternative hypothesis.

2] What is the null hypothesis, and how does it differ from the alternative hypothesis
- The null hypothesis is a statement that assumes there is no effect, difference, or relationship between variables. It represents the default or status quo. The alternative hypothesis is the opposite—it suggests that there is an effect or difference. In hypothesis testing, we test the null hypothesis and only reject it if the evidence strongly supports the alternative

3] Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
- The significance level, often denoted by alpha (α), is the threshold used to decide whether to reject the null hypothesis. Common values are 0.05 or 0.01. It represents the probability of making a Type I error—rejecting a true null hypothesis. It's important because it defines how strict the test is and helps control the risk of drawing incorrect conclusions from sample data.

4] What are Type I and Type II errors? Give examples of each.
- A Type I error occurs when you reject a true null hypothesis — it's a false positive.
Example: A medical test wrongly says a healthy person has a disease.

- A Type II error occurs when you fail to reject a false null hypothesis — it's a false negative.
Example: A medical test fails to detect a disease in a sick person.

5] What is the difference between a Z-test and a T-test. Explain when to use
each
- A Z-test is used when the population standard deviation is known and the sample size is large (typically ≥ 30). It assumes normality.

A T-test is used when the population standard deviation is unknown and the sample size is small (< 30). It uses the sample standard deviation and accounts for extra variability by using the t-distribution.

Use a Z-test for large samples with known variance; use a T-test for small samples or when variance is unknown.

```python
# 7] Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
# sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Parameters
mu_0 = 50  # hypothesized population mean
sigma = 0.5  # known population standard deviation
n = len(sample_data)
sample_mean = np.mean(sample_data)

# Z-test statistic
z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Sample mean = {sample_mean:.4f}")
print(f"Z-statistic = {z:.4f}")
print(f"P-value = {p_value:.4f}")
```

```python
# 8] Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate data
np.random.seed(0)
data = np.random.normal(loc=50, scale=5, size=100)

# Calculate sample mean and standard error
mean = np.mean(data)
sem = stats.sem(data)  # standard error of the mean

# 95% confidence interval using t-distribution
confidence = 0.95
ci = stats.t.interval(confidence, df=len(data)-1, loc=mean, scale=sem)

print(f"Mean = {mean:.2f}")
print(f"95% Confidence Interval = ({ci[0]:.2f}, {ci[1]:.2f})")

# Plot histogram and confidence interval
plt.hist(data, bins=15, color='skyblue', edgecolor='black')
plt.axvline(mean, color='red', linestyle='--', label='Mean')
plt.axvline(ci[0], color='green', linestyle='--', label='95% CI Lower')
plt.axvline(ci[1], color='green', linestyle='--', label='95% CI Upper')
plt.title("Normal Distribution Sample with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# 9] Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
import numpy as np
import matplotlib.pyplot as plt

def compute_and_plot_zscores(data):
    mean = np.mean(data)
    std = np.std(data)

    z_scores = (data - mean) / std

    print(f"Mean: {mean:.2f}, Std Dev: {std:.2f}")
    print("First 5 Z-scores:", z_scores[:5])

    # Plot
    plt.hist(z_scores, bins=15, color='lightcoral', edgecolor='black')
    plt.title("Histogram of Z-scores")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.axvline(0, color='blue', linestyle='--', label='Mean (Z=0)')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

    return z_scores

data = np.random.normal(100, 15, size=100)
compute_and_plot_zscores(data)

```
