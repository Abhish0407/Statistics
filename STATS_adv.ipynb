11] What is the measure of central tendency
- A measure of central tendency is a summary statistic that attempts to describe a whole set of data with a single value that represents the middle or center of its distribution.
There are three main measures of central tendency: Mean, Median, Mode.

12] Define mean, median, and mode
- Mean: Commonly referred to as the "average." It is calculated by summing all the values in a dataset and then dividing by the total number of values.

- Median: The median is the middle value in a dataset when the values are arranged in ascending or descending order. It divides the data into two equal halves.

- Mode: The mode is the value that appears most frequently in a dataset.

13] What is the significance of the measure of central tendency
- The measure of central tendency is significant because it provides a single value that represents the center or typical value of a dataset. It helps simplify complex data, making it easier to understand, compare, and analyze. It is widely used in decision-making, trend analysis, and interpreting statistical results.

14]  What is variance, and how is it calculated
- Variance measures how much the data values deviate from the mean. It shows the spread or dispersion of a dataset. To calculate it, subtract the mean from each data point, square the result, then take the average of those squared differences. For a population, divide by the number of data points (N); for a sample, divide by (N - 1).

15] What is standard deviation, and why is it important
- Standard deviation is the square root of variance. It shows how much data values deviate, on average, from the mean. It is important because it gives a clear sense of data spread in the same units as the data, making it easier to interpret variability and consistency within a dataset.

16] Define and explain the term range in statistics
- In statistics, the range is the difference between the highest and lowest values in a dataset. It shows how spread out the data is. For example, if the highest value is 90 and the lowest is 30, the range is 90 - 30 = 60. It's a simple measure of variability but doesn't reflect how data is distributed between the extremes.

17] What is the difference between variance and standard deviation
- Variance measures the average of the squared differences from the mean, while standard deviation is the square root of variance. Both show data spread, but standard deviation is in the same unit as the data, making it easier to interpret and compare.

18] What is skewness in a dataset
- Skewness measures the asymmetry of a dataset’s distribution.  

19] What does it mean if a dataset is positively or negatively skewed
- If the data is skewed to the right (positive skew), the tail is longer on the right side. If it's skewed to the left (negative skew), the tail is longer on the left.

20] Define and explain kurtosis
- Kurtosis measures the "tailedness" or peakedness of a data distribution. It indicates how much of the data is in the tails and center compared to a normal distribution.

21] What is the purpose of covariance
- The purpose of covariance is to measure how two variables change together. If both variables increase or decrease simultaneously, the covariance is positive. If one increases while the other decreases, it’s negative. It helps indicate the direction of the relationship between variables.

22] What is the purpose of correlation
- Correlation measures the strength and direction of a linear relationship between two variables. It tells how closely the variables move together. The value ranges from −1 to +1, where +1 indicates a perfect positive relationship, −1 a perfect negative relationship, and 0 means no linear relationship.

23] What is the difference between covariance and correlation
- Covariance measures the direction of the relationship between two variables, while correlation measures both the direction and strength of that relationship. Covariance is unscaled and depends on the units of the variables, making it harder to interpret. Correlation, on the other hand, is standardized and ranges from −1 to +1, making it easier to compare across different datasets.

```python
# 12] Implement and compare simple random sampling and systematic sampling
import pandas as pd
import numpy as np

# Create dataset
data = pd.DataFrame({'ID': np.arange(1, 101)})

# Simple Random Sample of 10 rows
simple_random_sample = data.sample(n=10, random_state=1)
print("Simple Random Sample:\n", simple_random_sample)


def systematic_sampling(df, step):
    start = np.random.randint(0, step)
    indexes = np.arange(start, len(df), step)
    return df.iloc[indexes]

# Systematic Sample (every 10th item)
systematic_sample = systematic_sampling(data, step=10)
print("\nSystematic Sample:\n", systematic_sample)
```

```python
# 13] Calculate the mean, median, and mode of grouped data
# Grouped data
classes = [(0, 10), (10, 20), (20, 30), (30, 40)]
frequencies = [5, 8, 15, 7]

# Midpoints
midpoints = [(a + b) / 2 for a, b in classes]

# Mean
total_freq = sum(frequencies)
mean = sum(f * m for f, m in zip(frequencies, midpoints)) / total_freq
print("Mean:", mean)

# Median
cumulative = [sum(frequencies[:i+1]) for i in range(len(frequencies))]
n = total_freq
for i, cf in enumerate(cumulative):
    if cf >= n / 2:
        median_class = i
        break

L = classes[median_class][0]
F = cumulative[median_class - 1] if median_class > 0 else 0
f = frequencies[median_class]
h = classes[median_class][1] - classes[median_class][0]
median = L + ((n/2 - F) / f) * h
print("Median:", median)

# Mode
modal = frequencies.index(max(frequencies))
L = classes[modal][0]
f1 = frequencies[modal]
f0 = frequencies[modal - 1] if modal > 0 else 0
f2 = frequencies[modal + 1] if modal < len(frequencies) - 1 else 0
mode = L + ((f1 - f0) / ((2*f1) - f0 - f2)) * h
print("Mode:", mode)
```

```python
# 3] Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): 0 or 1
    p (float): Probability of success (0 ≤ p ≤ 1)

    Returns:
    float: Probability of observing x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1")

    return p if x == 1 else 1 - p

print(bernoulli_pdf(1, 0.7))
```

```python
# 9] Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    """
    Calculate Z-scores and plot them.

    Parameters:
    data (list or array): Numeric dataset
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    z_scores = (data - mean) / std

    # Plotting
    plt.plot(z_scores, marker='o', linestyle='-', color='orange')
    plt.axhline(0, color='blue', linestyle='--')  # mean line
    plt.title("Z-scores of Dataset")
    plt.xlabel("Data Index")
    plt.ylabel("Z-score")
    plt.grid(True)
    plt.show()

    return z_scores

data = [10, 12, 9, 14, 13, 11, 15, 16, 10, 13]
z = calculate_and_plot_zscores(data)
print("Z-scores:", z)
```

```python
# 10] Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution
import numpy as np
import matplotlib.pyplot as plt

# Parameters
population_size = 100000
sample_size = 30
num_samples = 1000

# Step 1: Create a non-normal population (e.g., exponential)
population = np.random.exponential(scale=2.0, size=population_size)

# Step 2: Draw many sample means
sample_means = []
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size)
    sample_means.append(np.mean(sample))

# Step 3: Plot the distribution of sample means
plt.hist(sample_means, bins=30, density=True, color='green', edgecolor='black')
plt.title("Sampling Distribution of the Mean (CLT Demonstration)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
# 18] Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    """
    Calculate the Z-score of a data point.

    Parameters:
    x (float): Data point
    mean (float): Mean of the distribution
    std (float): Standard deviation of the distribution

    Returns:
    float: Z-score
    """
    return (x - mean) / std

# Example data
x_value = 72
mean = 70
std = 5

# Calculate Z-score
z = calculate_z_score(x_value, mean, std)
print(f"Z-score: {z:.2f}")

# Plot the standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.plot(x, y, label='Standard Normal PDF')
plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
plt.fill_between(x, 0, y, where=(x <= z), color='red', alpha=0.3)
plt.title('Z-score Position in Standard Normal Distribution')
plt.xlabel('Z')
plt.ylabel('Density')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 19] Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a one-sample Z-test.

    Parameters:
    sample_data (list or array): sample observations
    population_mean (float): hypothesized population mean
    population_std (float): known population standard deviation
    alpha (float): significance level

    Returns:
    None (prints result)
    """
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z:.2f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")

# Example usage
data = [68, 70, 72, 71, 69, 73, 75, 67, 70, 74]
pop_mean = 70
pop_std = 4  # Known population std

z_test(data, pop_mean, pop_std)
```

```python
# 20] Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate confidence interval for the mean of a dataset.

    Parameters:
    data (list or array): Sample data
    confidence (float): Confidence level (e.g., 0.95 for 95%)

    Returns:
    tuple: (lower_bound, upper_bound)
    """
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error
    margin = stats.t.ppf((1 + confidence) / 2, n - 1) * std_err

    lower = mean - margin
    upper = mean + margin

    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    return (lower, upper)

# Example usage
sample_data = [68, 70, 72, 71, 69, 73, 75, 67, 70, 74]
confidence_interval(sample_data, confidence=0.95)
```

```python
# 21] Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
from scipy import stats

# Step 1: Generate data from a normal distribution
np.random.seed(0)  # for reproducibility
mu = 100       # true mean
sigma = 15     # true standard deviation
n = 50         # sample size

data = np.random.normal(mu, sigma, n)

# Step 2: Calculate 95% confidence interval for the sample mean
sample_mean = np.mean(data)
std_err = stats.sem(data)
confidence = 0.95
margin = stats.t.ppf((1 + confidence) / 2, n - 1) * std_err
lower = sample_mean - margin
upper = sample_mean + margin

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
# 23] Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters for the Poisson distribution
mu = 4  # average rate (λ)

# Generate possible values of x
x = np.arange(0, 15)

# Calculate the CDF values
cdf_values = poisson.cdf(x, mu)

# Display CDF values
for xi, cdf in zip(x, cdf_values):
    print(f"P(X ≤ {xi}) = {cdf:.4f}")

# Plot the CDF
plt.step(x, cdf_values, where='post', label=f'Poisson(λ={mu}) CDF', color='purple')
plt.title("Cumulative Distribution Function (CDF) of Poisson Distribution")
plt.xlabel("x (Number of Events)")
plt.ylabel("P(X ≤ x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 28] Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution
import numpy as np
from scipy.stats import skew, kurtosis

def compute_skewness_kurtosis(mean=0, std=1, size=1000):
    """
    Generate data from a normal distribution and compute skewness and kurtosis.

    Parameters:
    mean (float): Mean of the normal distribution
    std (float): Standard deviation
    size (int): Number of samples

    Returns:
    tuple: (skewness, kurtosis)
    """
    # Generate normal data
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness and kurtosis
    skew_val = skew(data)
    kurt_val = kurtosis(data)  # excess kurtosis (normal = 0)

    print(f"Skewness: {skew_val:.4f}")
    print(f"Kurtosis (excess): {kurt_val:.4f}")

    return skew_val, kurt_val

# Example usage
compute_skewness_kurtosis()
```
