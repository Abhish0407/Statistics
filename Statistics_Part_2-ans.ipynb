1] What is hypothesis testing in statistics
- Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It involves forming a null hypothesis (usually stating no effect or difference) and an alternative hypothesis.

2] What is the null hypothesis, and how does it differ from the alternative hypothesis
- The null hypothesis is a statement that assumes there is no effect, difference, or relationship between variables. It represents the default or status quo. The alternative hypothesis is the opposite—it suggests that there is an effect or difference. In hypothesis testing, we test the null hypothesis and only reject it if the evidence strongly supports the alternative

3] What is the significance level in hypothesis testing, and why is it important
- The significance level, often denoted by alpha (α), is the threshold used to decide whether to reject the null hypothesis. Common values are 0.05 or 0.01. It represents the probability of making a Type I error—rejecting a true null hypothesis. It's important because it defines how strict the test is and helps control the risk of drawing incorrect conclusions from sample data.

4] What does a P-value represent in hypothesis testing
- A p-value represents the probability of obtaining results as extreme as, or more extreme than, the observed results, assuming the null hypothesis is true. A small p-value (typically ≤ 0.05) indicates strong evidence against the null hypothesis, suggesting it should be rejected. A large p-value suggests weak evidence, so the null hypothesis is not rejected.

5] How do you interpret the P-value in hypothesis testing
- A p-value represents the probability of obtaining results as extreme as, or more extreme than, the observed results, assuming the null hypothesis is true. A small p-value (typically ≤ 0.05) indicates strong evidence against the null hypothesis, suggesting it should be rejected. A large p-value suggests weak evidence, so the null hypothesis is not rejected.

6] What are Type 1 and Type 2 errors in hypothesis testing
- Type 1 error occurs when the null hypothesis is true but is wrongly rejected (a false positive). Type 2 error happens when the null hypothesis is false but is wrongly accepted (a false negative). In simple terms, Type 1 is detecting an effect that isn’t there, while Type 2 is missing an effect that actually exists.

7] What is the difference between a one-tailed and a two-tailed test in hypothesis testing
- A one-tailed test checks for an effect in only one direction (either greater than or less than), while a two-tailed test checks for an effect in both directions (not equal to). Use a one-tailed test when you're only interested in change in one direction, and a two-tailed test when any significant difference matters, regardless of direction

8] What is the Z-test, and when is it used in hypothesis testing
- A Z-test is a statistical test used to determine whether there is a significant difference between sample and population means when the population standard deviation is known. It is typically used when the sample size is large (n ≥ 30) and the data is approximately normally distributed

9] How do you calculate the Z-score, and what does it represent in hypothesis testing
- The Z-score is calculated using the formula:
Z = (X - μ) / σ,
where X is the sample value, μ is the population mean, and σ is the population standard deviation.

It represents how many standard deviations a data point is from the mean. In hypothesis testing, it helps determine how extreme a sample result is under the assumption that the null hypothesis is true.

10] What is the T-distribution, and when should it be used instead of the normal distribution
- The t-distribution is a probability distribution used when estimating population parameters with a small sample size (typically n < 30) and the population standard deviation is unknown. It is similar to the normal distribution but has heavier tails, which provide more accurate results for small samples. As the sample size increases, the t-distribution approaches the normal distribution.

11] What is the difference between a Z-test and a T-test
- A Z-test is used when the sample size is large (n ≥ 30) and the population standard deviation is known. A T-test is used when the sample size is small (n < 30) and the population standard deviation is unknown. The T-test accounts for more uncertainty with small samples, using the t-distribution instead of the normal distribution.

12] What is the T-test, and how is it used in hypothesis testing
- A T-test is a statistical test used to compare the means of two groups when the population standard deviation is unknown. It helps determine whether the observed difference between group means is statistically significant.

13] What is the relationship between Z-test and T-test in hypothesis testing
- The Z-test and T-test both assess whether a sample mean differs significantly from a population mean or another sample mean. They use similar logic and formulas, but the key difference is that the Z-test is used when the population standard deviation is known and the sample size is large, while the T-test is used when the population standard deviation is unknown and the sample size is small.

14] What is a confidence interval, and how is it used to interpret statistical results
- A confidence interval is a range of values, derived from sample data, that likely contains the true population parameter. It is used to express the uncertainty around a sample estimate. For example, a 95% confidence interval means that if the same population were sampled many times, 95% of the intervals would contain the true value. It helps interpret results by showing not just a single estimate but the possible range of values with a given level of confidence.

15] What is the margin of error, and how does it affect the confidence interval
- The margin of error is the amount added and subtracted from a sample estimate to create a confidence interval. It reflects the uncertainty or possible error in the estimate due to sampling variability. A larger margin of error results in a wider confidence interval, indicating less precision, while a smaller margin of error gives a narrower interval, showing more precise estimates. It is influenced by sample size, variability, and confidence level

16] How is Bayes' Theorem used in statistics, and what is its significance
- Bayes' Theorem is used in statistics to update the probability of a hypothesis based on new evidence. It combines prior knowledge (prior probability) with observed data (likelihood) to calculate a revised probability (posterior probability). Its significance lies in allowing probabilistic reasoning and decision-making under uncertainty, making it widely used in fields like machine learning, medicine, and finance.

17] What is the Chi-square distribution, and when is it used
- The Chi-square distribution is a probability distribution used primarily for testing relationships between categorical variables. It is used in statistical tests like the Chi-square test for independence and the goodness-of-fit test. This distribution is skewed right and depends on degrees of freedom. It's most commonly applied when assessing whether observed frequencies differ significantly from expected frequencies.

18] What is the Chi-square goodness of fit test, and how is it applied
- The Chi-square goodness of fit test is used to determine whether the observed frequencies in categories match the expected frequencies based on a specific distribution. It compares actual data to a theoretical model to see if the data fits well. To apply it, calculate the Chi-square statistic using the formula:
χ² = Σ[(O − E)² / E],
where O is observed frequency and E is expected frequency. Then, compare the result to a critical value from the Chi-square distribution to decide whether to reject the null hypothesis that the data fits the expected distribution.

19] What is the F-distribution, and when is it used in hypothesis testing
- The F-distribution is a probability distribution used mainly to compare variances between two groups. It is right-skewed and depends on two sets of degrees of freedom. In hypothesis testing, it is used in ANOVA (Analysis of Variance) to test whether three or more group means are significantly different, and in F-tests to compare the variances of two populations

20] What is an ANOVA test, and what are its assumptions
- ANOVA (Analysis of Variance) is a statistical test used to determine whether there are significant differences between the means of three or more groups. It compares the variance within groups to the variance between groups to see if at least one group mean is different from the others.

The main assumptions of ANOVA are:

 - The samples are independent.

 - The data in each group are normally distributed.

 - All groups have equal variances (homogeneity of variance).

21] What are the different types of ANOVA tests
- There are three main types of ANOVA tests:

 - One-Way ANOVA – Tests differences between means of three or more independent groups based on one factor (e.g., comparing test scores across different classes).

 - Two-Way ANOVA – Examines the effect of two independent variables on a dependent variable, and also checks for interaction between them (e.g., studying the effect of teaching method and gender on scores).

 - Repeated Measures ANOVA – Used when the same subjects are tested under multiple conditions or over time (e.g., measuring blood pressure before, during, and after treatment).

22] What is the F-test, and how does it relate to hypothesis testing
- The F-test is a statistical test used to compare two variances or to test whether group means are significantly different, especially in ANOVA. It involves calculating an F-statistic, which is the ratio of variances between groups to the variance within groups. In hypothesis testing, if the F-value is significantly large, it suggests that at least one group mean differs, leading to rejection of the null hypothesis that all group means are equal.

```python
# 3] Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
import numpy as np
import matplotlib.pyplot as plt

def bernoulli_pdf(p, x):
    """
    Calculate the probability density function (PDF) of a Bernoulli distribution.

    Parameters:
    - p (float): Probability of success (1), must be between 0 and 1
    - x (int or array-like): Value(s) at which to evaluate the PDF (0 or 1)

    Returns:
    - float or ndarray: PDF value(s) at x
    """
    x = np.asarray(x)
    if not np.all(np.isin(x, [0, 1])):
        raise ValueError("x must be 0 or 1 for Bernoulli distribution")
    if not 0 <= p <= 1:
        raise ValueError("p must be between 0 and 1")

    return p**x * (1 - p)**(1 - x)

# Example usage:
p = 0.7  # Probability of success
x_values = [0, 1]  # Possible outcomes

# Calculate PDF
pdf_values = bernoulli_pdf(p, x_values)

# Plot the distribution
plt.stem(x_values, pdf_values, basefmt=" ")
plt.xticks([0, 1])
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title(f'Bernoulli Distribution PDF (p={p})')
plt.grid(True, alpha=0.3)
plt.show()

print(f"PDF at x=0: {bernoulli_pdf(p, 0):.2f}")
print(f"PDF at x=1: {bernoulli_pdf(p, 1):.2f}")
```

```python
# 6] Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

def plot_discrete_uniform_cdf(low, high):
    """
    Calculate and plot the CDF of a discrete uniform distribution

    Parameters:
    low (int): Minimum value (inclusive)
    high (int): Maximum value (exclusive)
    """
    # Create the distribution object
    rv = randint(low, high)

    # Generate possible values
    x = np.arange(low - 1, high + 1)  # Extend range for proper step visualization
    cdf_values = rv.cdf(x)

    # Plot CDF
    plt.figure(figsize=(10, 6))
    plt.step(x, cdf_values, where='post', linewidth=2.5)

    # Mark the discrete points
    plt.plot(x[1:-1], cdf_values[1:-1], 'bo', markersize=8)

    # Formatting
    plt.title(f'CDF of Discrete Uniform Distribution (low={low}, high={high-1})', pad=20)
    plt.xlabel('Value')
    plt.ylabel('Cumulative Probability')
    plt.xticks(np.arange(low - 1, high + 1))
    plt.yticks(np.linspace(0, 1, 11))
    plt.grid(True, alpha=0.3)
    plt.ylim(-0.05, 1.05)

    # Highlight key properties
    plt.axhline(y=1, color='gray', linestyle='--', alpha=0.5)
    plt.axvline(x=low, color='red', linestyle=':', alpha=0.5)
    plt.axvline(x=high-1, color='red', linestyle=':', alpha=0.5)

    plt.show()

# Example usage:
plot_discrete_uniform_cdf(low=1, high=7)  # Values 1 through 6
```

```python
# 9] Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore, norm

def plot_z_scores(data, threshold=2.0):
    """
    Calculate Z-scores and create a visualization with histogram, density plot, and outliers marked.

    Parameters:
    - data: Input array or list
    - threshold: Z-score threshold for marking outliers (default: 2.0)
    """
    # Convert to numpy array if not already
    data = np.asarray(data)

    # Calculate Z-scores
    z_scores = zscore(data)

    # Identify outliers
    outliers = np.abs(z_scores) > threshold

    # Create figure with 3 subplots
    plt.figure(figsize=(15, 5))

    # Subplot 1: Original data with outliers marked
    plt.subplot(1, 3, 1)
    plt.scatter(range(len(data)), data, c=outliers, cmap='cool', alpha=0.7)
    plt.axhline(np.mean(data), color='k', linestyle='--', label=f'Mean: {np.mean(data):.2f}')
    plt.title('Original Data with Outliers')
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(alpha=0.3)

    # Subplot 2: Z-scores distribution
    plt.subplot(1, 3, 2)
    plt.hist(z_scores, bins=30, density=True, alpha=0.7, color='orange', edgecolor='black')

    # Plot standard normal distribution for comparison
    x = np.linspace(-4, 4, 100)
    plt.plot(x, norm.pdf(x), 'r-', label='Standard Normal')
    plt.axvline(threshold, color='orange', linestyle=':', label=f'Threshold: ±{threshold}')
    plt.axvline(-threshold, color='orange', linestyle=':')
    plt.title('Z-scores Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(alpha=0.3)

    # Subplot 3: Q-Q plot for normality check
    plt.subplot(1, 3, 3)
    norm_quantiles = np.percentile(norm.rvs(size=10000), np.linspace(0, 100, len(data)))
    plt.scatter(norm_quantiles, np.sort(z_scores), alpha=0.7)
    plt.plot([-3, 3], [-3, 3], 'r--', label='Perfect normality')
    plt.title('Q-Q Plot (Normality Check)')
    plt.xlabel('Theoretical Quantiles')
    plt.ylabel('Sample Z-scores Quantiles')
    plt.legend()
    plt.grid(alpha=0.3)

    plt.tight_layout()
    plt.show()

    # Return both original Z-scores and outlier flags
    return z_scores, outliers

# Example usage:
if __name__ == "__main__":
    # Create sample data with some outliers
    np.random.seed(42)
    data = np.concatenate([np.random.normal(0, 1, 95),
                          np.random.normal(5, 1, 5)])

    z_scores, outliers = plot_z_scores(data, threshold=2.0)

    # Print some statistics
    print(f"\n{'='*50}")
    print(f"{'Z-score Analysis':^50}")
    print(f"{'='*50}")
    print(f"Data points: {len(data)}")
    print(f"Mean Z-score: {np.mean(z_scores):.4f}")
    print(f"Std Dev of Z-scores: {np.std(z_scores):.4f}")
    print(f"Outliers detected: {np.sum(outliers)} (Z > ±{2.0})")
    print(f"Outlier percentage: {np.mean(outliers)*100:.2f}%")
    print(f"Largest Z-score: {np.max(np.abs(z_scores)):.2f}")
```

```python
# 10] Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon, norm

def demonstrate_clt(sample_size=30, num_samples=1000, lambda_param=1.0):
    """
    Demonstrate the Central Limit Theorem using an exponential distribution.

    Parameters:
    - sample_size: Size of each sample (n)
    - num_samples: Number of samples to draw
    - lambda_param: Rate parameter for exponential distribution
    """
    # Set random seed for reproducibility
    np.random.seed(42)

    # Generate samples from exponential distribution
    population = expon(scale=1/lambda_param)
    samples = population.rvs(size=(num_samples, sample_size))

    # Calculate sample means
    sample_means = np.mean(samples, axis=1)

    # Create figure
    plt.figure(figsize=(15, 5))

    # Plot 1: Original exponential distribution
    plt.subplot(1, 3, 1)
    x = np.linspace(0, 5/lambda_param, 1000)
    plt.plot(x, population.pdf(x), 'r-', lw=2)
    plt.title('Original Exponential Distribution\n(λ={})'.format(lambda_param))
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.grid(alpha=0.3)

    # Plot 2: Histogram of sample means
    plt.subplot(1, 3, 2)
    plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')

    # Overlay theoretical normal distribution
    theo_mean = 1/lambda_param  # Mean of exponential distribution
    theo_std = (1/lambda_param)/np.sqrt(sample_size)  # Standard error

    x_norm = np.linspace(theo_mean - 3*theo_std, theo_mean + 3*theo_std, 100)
    plt.plot(x_norm, norm.pdf(x_norm, theo_mean, theo_std), 'r-', lw=2)
    plt.title('Distribution of Sample Means\n(n={})'.format(sample_size))
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.grid(alpha=0.3)

    # Plot 3: Q-Q plot of sample means
    plt.subplot(1, 3, 3)
    norm_quantiles = norm.ppf(np.linspace(0.01, 0.99, len(sample_means)))
    sample_quantiles = np.sort(sample_means)
    plt.scatter(norm_quantiles, sample_quantiles, alpha=0.7)

    # Perfect normality line
    min_val = min(norm_quantiles.min(), sample_quantiles.min())
    max_val = max(norm_quantiles.max(), sample_quantiles.max())
    plt.plot([min_val, max_val], [min_val, max_val], 'r--')
    plt.title('Q-Q Plot of Sample Means')
    plt.xlabel('Theoretical Normal Quantiles')
    plt.ylabel('Sample Quantiles')
    plt.grid(alpha=0.3)

    plt.tight_layout()
    plt.show()

    # Print statistics
    print(f"\n{'='*60}")
    print(f"{'Central Limit Theorem Demonstration':^60}")
    print(f"{'='*60}")
    print(f"{'Population mean (theoretical)':<40}: {theo_mean:.4f}")
    print(f"{'Population mean (simulated)':<40}: {np.mean(samples):.4f}")
    print(f"{'Mean of sample means':<40}: {np.mean(sample_means):.4f}")
    print(f"{'Standard error (theoretical)':<40}: {theo_std:.4f}")
    print(f"{'Standard deviation of sample means':<40}: {np.std(sample_means):.4f}")
    print(f"Sample size (n): {sample_size}")
    print(f"Number of samples: {num_samples}")

# Run the demonstration
demonstrate_clt(sample_size=30, num_samples=1000, lambda_param=1.0)
```

```python
# 11] Simulate multiple samples from a normal distribution and verify the Central Limit Theorem
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, skew, kurtosis

def verify_clt(pop_mean=5, pop_std=2, sample_size=30, num_samples=1000):
    """
    Verify the Central Limit Theorem by sampling from a normal distribution.

    Parameters:
    - pop_mean: Population mean (μ)
    - pop_std: Population standard deviation (σ)
    - sample_size: Size of each sample (n)
    - num_samples: Number of samples to draw
    """
    np.random.seed(42)  # For reproducibility

    # 1. Generate population data (normal distribution)
    population = norm(loc=pop_mean, scale=pop_std)

    # 2. Draw multiple samples
    samples = population.rvs(size=(num_samples, sample_size))

    # 3. Calculate sample means
    sample_means = np.mean(samples, axis=1)

    # 4. Theoretical predictions from CLT
    theo_mean = pop_mean
    theo_std = pop_std / np.sqrt(sample_size)

    # Create visualization
    plt.figure(figsize=(15, 5))

    # Plot 1: Population distribution
    plt.subplot(1, 3, 1)
    x_pop = np.linspace(pop_mean - 4*pop_std, pop_mean + 4*pop_std, 1000)
    plt.plot(x_pop, population.pdf(x_pop), 'r-', label=f'N(μ={pop_mean}, σ={pop_std})')
    plt.title('Population Distribution')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(alpha=0.3)

    # Plot 2: Distribution of sample means
    plt.subplot(1, 3, 2)
    plt.hist(sample_means, bins=30, density=True, alpha=0.7,
             color='skyblue', edgecolor='black')

    # Overlay theoretical normal distribution
    x_means = np.linspace(theo_mean - 4*theo_std, theo_mean + 4*theo_std, 1000)
    plt.plot(x_means, norm.pdf(x_means, theo_mean, theo_std),
             'r-', label=f'N(μ={theo_mean:.2f}, σ={theo_std:.2f})')
    plt.title(f'Distribution of Sample Means\n(n={sample_size})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(alpha=0.3)

    # Plot 3: Q-Q plot for normality check
    plt.subplot(1, 3, 3)
    sample_means_sorted = np.sort(sample_means)
    theoretical_quantiles = norm.ppf(np.linspace(0.01, 0.99, len(sample_means)))
    plt.scatter(theoretical_quantiles, sample_means_sorted, alpha=0.7)

    # Perfect normality line
    min_val = min(theoretical_quantiles.min(), sample_means_sorted.min())
    max_val = max(theoretical_quantiles.max(), sample_means_sorted.max())
    plt.plot([min_val, max_val], [min_val, max_val], 'r--')
    plt.title('Q-Q Plot of Sample Means')
    plt.xlabel('Theoretical Normal Quantiles')
    plt.ylabel('Sample Quantiles')
    plt.grid(alpha=0.3)

    plt.tight_layout()
    plt.show()

    # Calculate statistics
    actual_mean = np.mean(sample_means)
    actual_std = np.std(sample_means)
    skewness = skew(sample_means)
    kurt = kurtosis(sample_means)

    # Print verification results
    print(f"\n{' CLT Verification Results ':=^60}")
    print(f"{'Theoretical mean of sample means':<40}: {theo_mean:.4f}")
    print(f"{'Actual mean of sample means':<40}: {actual_mean:.4f}")
    print(f"\n{'Theoretical standard error (σ/√n)':<40}: {theo_std:.4f}")
    print(f"{'Actual standard deviation of means':<40}: {actual_std:.4f}")
    print(f"\n{'Skewness of sample means':<40}: {skewness:.4f} (0 is normal)")
    print(f"{'Kurtosis of sample means':<40}: {kurt:.4f} (0 is normal)")
    print(f"\n{'95% of means should be within':<40}: {theo_mean - 1.96*theo_std:.2f} to {theo_mean + 1.96*theo_std:.2f}")
    print(f"{'Actual percentage in this range':<40}: {np.mean((sample_means > theo_mean - 1.96*theo_std) & (sample_means < theo_mean + 1.96*theo_std))*100:.2f}%")

# Run the verification
verify_clt(pop_mean=5, pop_std=2, sample_size=30, num_samples=1000)
```

```python
# 13] Generate random variables and calculate their corresponding probabilities using the binomial distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def binomial_distribution_analysis(n=10, p=0.5, num_samples=1000):
    """
    Generate binomial random variables, calculate probabilities, and visualize results.

    Parameters:
    n (int): Number of trials
    p (float): Probability of success (0 ≤ p ≤ 1)
    num_samples (int): Number of random variables to generate
    """
    # Generate random variables
    random_vars = np.random.binomial(n, p, num_samples)

    # Calculate theoretical probabilities
    k_values = np.arange(0, n+1)
    pmf = binom.pmf(k_values, n, p)

    # Create figure
    plt.figure(figsize=(12, 5))

    # Plot 1: Histogram of generated random variables
    plt.subplot(1, 2, 1)
    plt.hist(random_vars, bins=np.arange(-0.5, n+1.5, 1),
             density=True, alpha=0.7, color='skyblue', edgecolor='black')
    plt.stem(k_values, pmf, linefmt='r-', markerfmt='ro', basefmt=' ',
             label='Theoretical PMF')
    plt.title(f'Binomial Distribution (n={n}, p={p})\nSimulation vs Theory')
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.xticks(k_values)
    plt.legend()
    plt.grid(alpha=0.3)

    # Plot 2: Cumulative distribution
    plt.subplot(1, 2, 2)
    cdf = binom.cdf(k_values, n, p)
    plt.step(k_values, cdf, where='post', label='Theoretical CDF')
    plt.title('Cumulative Distribution Function')
    plt.xlabel('Number of Successes')
    plt.ylabel('Cumulative Probability')
    plt.xticks(k_values)
    plt.grid(alpha=0.3)

    plt.tight_layout()
    plt.show()

    # Print probability table
    print(f"\n{' Binomial Distribution Analysis ':=^60}")
    print(f"{'Parameters':<20}: n={n}, p={p}")
    print(f"{'Mean (np)':<20}: {n*p:.2f}")
    print(f"{'Variance (np(1-p))':<20}: {n*p*(1-p):.2f}\n")

    print(f"{'k':<5}{'P(X=k)':<15}{'P(X≤k)':<15}{'P(X≥k)':<15}")
    print("-"*45)
    for k in k_values:
        prob_eq = binom.pmf(k, n, p)
        prob_leq = binom.cdf(k, n, p)
        prob_geq = 1 - binom.cdf(k-1, n, p) if k > 0 else 1
        print(f"{k:<5}{prob_eq:<15.4f}{prob_leq:<15.4f}{prob_geq:<15.4f}")

# Example usage
binomial_distribution_analysis(n=10, p=0.3, num_samples=5000)
```

```python
# 14] Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def analyze_z_score(data_point, data_mean, data_std):
    """
    Calculate the Z-score for a data point and visualize its position in the standard normal distribution.

    Parameters:
    - data_point: The value to analyze
    - data_mean: Mean of the dataset
    - data_std: Standard deviation of the dataset
    """
    # Calculate Z-score
    z_score = (data_point - data_mean) / data_std

    # Create figure
    plt.figure(figsize=(12, 6))

    # Generate standard normal curve
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    # Plot standard normal distribution
    plt.plot(x, y, 'b-', label='Standard Normal Distribution (μ=0, σ=1)')

    # Highlight Z-score position
    plt.axvline(z_score, color='r', linestyle='--',
                label=f'Data Point (Z = {z_score:.2f})')

    # Fill relevant areas
    if z_score > 0:
        fill_start = -4
        fill_end = z_score
    else:
        fill_start = z_score
        fill_end = 4

    plt.fill_between(x, y, where=(x >= fill_start) & (x <= fill_end),
                    color='skyblue', alpha=0.3)

    # Add probability annotations
    prob_left = norm.cdf(z_score)
    prob_right = 1 - prob_left
    plt.text(z_score, norm.pdf(z_score)+0.02,
             f'P(Z ≤ {z_score:.2f}) = {prob_left:.2%}',
             ha='center', va='bottom')
    plt.text(z_score, norm.pdf(z_score)-0.02,
             f'P(Z > {z_score:.2f}) = {prob_right:.2%}',
             ha='center', va='top')

    # Add standard deviation markers
    for i in range(-3, 4):
        if i != 0:
            plt.axvline(i, color='gray', linestyle=':', alpha=0.5)

    # Format plot
    plt.title(f'Z-Score Analysis (Value = {data_point}, μ = {data_mean}, σ = {data_std})')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.ylim(0, 0.45)

    plt.tight_layout()
    plt.show()

    # Print interpretation
    print(f"\n{' Z-Score Analysis ':=^60}")
    print(f"{'Original value':<25}: {data_point}")
    print(f"{'Dataset mean':<25}: {data_mean}")
    print(f"{'Dataset standard deviation':<25}: {data_std}")
    print(f"{'Calculated Z-score':<25}: {z_score:.2f}")

    print("\nInterpretation:")
    if abs(z_score) < 1:
        print("- This is a typical value (within 1 standard deviation)")
    elif abs(z_score) < 2:
        print("- This is an unusual value (1-2 standard deviations)")
    elif abs(z_score) < 3:
        print("- This is a very unusual value (2-3 standard deviations)")
    else:
        print("- This is an extremely unusual value (>3 standard deviations)")

    print(f"\nProbability Analysis:")
    print(f"- Probability of values ≤ this: {prob_left:.2%}")
    print(f"- Probability of values > this: {prob_right:.2%}")

# Example usage
if __name__ == "__main__":
    # Example dataset parameters
    dataset_mean = 50
    dataset_std = 10

    # Data point to analyze
    test_value = 65  # Try changing this value

    analyze_z_score(test_value, dataset_mean, dataset_std)
```

```python
# 4]  Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
population_mean = 100
population_std = 15
sample_size = 50
alpha = 0.05

# Simulate sample and compute statistics
np.random.seed(0)
sample = np.random.normal(loc=105, scale=population_std, size=sample_size)
sample_mean = np.mean(sample)
z = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Critical z-values for two-tailed test
z_critical = norm.ppf(1 - alpha / 2)

# Plot
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal Curve')

# Shade rejection regions
plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3, label='Rejection Region')

# Plot observed z-score
plt.axvline(z, color='blue', linestyle='--', label=f'Observed z = {z:.2f}')
plt.axvline(-z_critical, color='black', linestyle='--', label='Critical z = ±{:.2f}'.format(z_critical))
plt.axvline(z_critical, color='black', linestyle='--')

plt.title('Two-tailed Z-test Decision Region')
plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 5] Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=0, mu1=1, sigma=1, alpha=0.05, n=30):
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value under H0 (two-tailed)
    z_crit = norm.ppf(1 - alpha / 2)
    crit_low = mu0 - z_crit * se
    crit_high = mu0 + z_crit * se

    # Plotting range
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # PDFs under H0 and H1
    y0 = norm.pdf(x, mu0, se)  # Null distribution
    y1 = norm.pdf(x, mu1, se)  # Alternative distribution

    # Plot
    plt.figure(figsize=(10, 5))
    plt.plot(x, y0, label='H0: μ = {}'.format(mu0), color='blue')
    plt.plot(x, y1, label='H1: μ = {}'.format(mu1), color='green')

    # Type I Error: Reject H0 when it's true (alpha)
    plt.fill_between(x, y0, where=(x < crit_low) | (x > crit_high), color='red', alpha=0.3, label='Type I Error (α)')

    # Type II Error: Fail to reject H0 when H1 is true (beta)
    plt.fill_between(x, y1, where=(x > crit_low) & (x < crit_high), color='orange', alpha=0.3, label='Type II Error (β)')

    # Critical values
    plt.axvline(crit_low, color='black', linestyle='--')
    plt.axvline(crit_high, color='black', linestyle='--')

    plt.title('Type I and Type II Errors Visualization')
    plt.xlabel('Test Statistic')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
visualize_type1_type2_errors(mu0=0, mu1=1, sigma=1, alpha=0.05, n=30)
```

```python
# 6] Write a Python program to perform an independent T-test and interpret the results
import numpy as np
from scipy.stats import ttest_ind

# Simulate two independent samples
np.random.seed(0)
group1 = np.random.normal(loc=100, scale=10, size=50)
group2 = np.random.normal(loc=105, scale=10, size=50)

# Perform independent t-test
t_stat, p_value = ttest_ind(group1, group2)

# Output results
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis: Means are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in means.")
```

```python
# 8] Simulate data and perform both Z-test and T-test, then compare the results using Python
import numpy as np
from scipy.stats import ttest_1samp, norm

# Simulate sample data
np.random.seed(1)
data = np.random.normal(loc=102, scale=15, size=30)  # sample size = 30
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
population_mean = 100

# Z-test
z_score = (sample_mean - population_mean) / (15 / np.sqrt(len(data)))
z_p_value = 2 * (1 - norm.cdf(abs(z_score)))

# T-test
t_stat, t_p_value = ttest_1samp(data, popmean=population_mean)

# Output
print(f"Z-test: z = {z_score:.4f}, p = {z_p_value:.4f}")
print(f"T-test: t = {t_stat:.4f}, p = {t_p_value:.4f}")

if z_p_value < 0.05:
    print("Z-test: Reject null hypothesis")
else:
    print("Z-test: Fail to reject null hypothesis")

if t_p_value < 0.05:
    print("T-test: Reject null hypothesis")
else:
    print("T-test: Fail to reject null hypothesis")
```

```python
# 9] Write a Python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    sem = np.std(data, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    margin = t_crit * sem
    lower = mean - margin
    upper = mean + margin
    return (lower, upper)

# Example usage
sample_data = [12, 15, 14, 10, 13, 12, 14, 13]
ci = confidence_interval(sample_data)
print(f"95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")
```

```python
# 10]  Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):
    n = len(data)
    sample_mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)
    moe = t_critical * std_err
    return moe

# Example usage
sample_data = [45, 50, 55, 52, 47, 53, 49, 51, 48, 54]
moe = margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error (95% confidence): {moe:.2f}")
```

```python
# 11] Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
# Bayesian Inference using Bayes' Theorem

def bayes_theorem(prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A):
    # Bayes' Theorem
    numerator = likelihood_B_given_A * prior_A
    denominator = (likelihood_B_given_A * prior_A) + (likelihood_B_given_not_A * prior_not_A)
    posterior_A_given_B = numerator / denominator
    return posterior_A_given_B

# Example:
# Disease test:
# P(Disease) = 0.01
# P(No Disease) = 0.99
# P(Positive | Disease) = 0.99 (true positive rate)
# P(Positive | No Disease) = 0.05 (false positive rate)

P_D = 0.01
P_not_D = 0.99
P_Pos_given_D = 0.99
P_Pos_given_not_D = 0.05

posterior = bayes_theorem(P_D, P_Pos_given_D, P_not_D, P_Pos_given_not_D)
print(f"Probability of having disease given a positive test: {posterior:.4f}")
```

```python
# 12] Perform a Chi-square test for independence between two categorical variables in Python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table
# Example: survey on gender and preference
data = {
    'Likes Coffee': [30, 10],
    'Likes Tea': [20, 40]
}
table = pd.DataFrame(data, index=['Male', 'Female'])

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(table)

# Output results
print("Chi-square Statistic:", round(chi2, 4))
print("Degrees of Freedom:", dof)
print("P-value:", round(p, 4))
print("Expected Frequencies:\n", np.round(expected, 2))

# Interpretation
if p < 0.05:
    print("Reject the null hypothesis: There is a significant association between the variables.")
else:
    print("Fail to reject the null hypothesis: No significant association between the variables.")
```

```python
# 14] Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution
import numpy as np
from scipy.stats import chisquare

# Observed frequencies (e.g., dice rolls)
observed = np.array([18, 22, 20, 17, 21, 22])

# Expected frequencies (uniform distribution for a fair 6-sided die)
expected = np.array([20, 20, 20, 20, 20, 20])

# Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Output
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis: Observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: Observed data fits the expected distribution.")
```
