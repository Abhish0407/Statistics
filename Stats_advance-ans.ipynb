1] What is a random variable in probability theory
- A random variable is a function that assigns a numerical value to each outcome in a probability space, representing uncertain events or quantities.

2] What are the types of random variables
- Random variables are mainly two types:

 - Discrete: Takes finite or countable values (e.g., number of heads in coin flips).
 - Continuous: Takes values in a continuous range (e.g., height of a person).

3] What is the difference between discrete and continuous distributions
- Discrete takes finite or countable values whereas Continuous takes values in a continuous range.

4] What are probability distribution functions (PDF)
- A Probability Distribution Function (PDF) defines the likelihood of a continuous random variable taking on a range of values.

5]  How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)
- A PDF shows the likelihood of a continuous variable taking on a specific value range, while a CDF gives the probability that the variable is less than or equal to a certain value. The CDF is the cumulative sum (integral) of the PDF up to that point.

6] What is a discrete uniform distribution
- A discrete uniform distribution is a probability distribution where a discrete random variable has a finite number of equally likely outcomes. Each outcome has the same probability, typically {1}/{n}, where n  is the number of possible values.

7] What are the key properties of a Bernoulli distribution
- A Bernoulli distribution models a single trial with two outcomes (success or failure). Key properties:

 - Outcomes: Success (1) with probability $ p $, failure (0) with probability 1-p .
 - Probability Mass Function (PMF): P(X = x) = p^x (1-p)^{1-x} , where  x = 0  or  1 .
 - Mean:  E[X] = p .
 - Variance:  Var(X) = p(1-p) .
 - Range:  X \in \{0, 1\} .
 - Single trial: Represents one binary event (e.g., coin flip).


8] What is the binomial distribution, and how is it used in probability
- The binomial distribution models the number of successes in a fixed number of independent trials, where each trial has only two outcomes (success or failure) and a constant probability of success. It's used to calculate probabilities in situations like flipping coins, passing/failing tests, or any yes/no outcomes repeated multiple times.

9] What is the Poisson distribution and where is it applied
- The Poisson distribution models the number of times an event occurs in a fixed interval of time or space, assuming the events happen independently and at a constant average rate. It's commonly applied in scenarios like counting phone calls per hour at a call center, arrivals at a bus stop, or defects per meter in manufacturing

10] What is a continuous uniform distribution
- A continuous uniform distribution describes a situation where all values within a given range are equally likely to occur. The probability is spread evenly across the interval, and its PDF is flat. An example is randomly picking a number between 0 and 1 where every value has the same chance.

11] What are the characteristics of a normal distribution
- A normal distribution is symmetric and bell-shaped, with most data centered around the mean. The mean, median, and mode are all equal, and about 68% of values lie within one standard deviation, 95% within two, and 99.7% within three. It’s widely used in statistics because many natural phenomena follow this pattern

12] What is the standard normal distribution, and why is it important
- The standard normal distribution is a special case of the normal distribution with a mean of 0 and a standard deviation of 1. It’s important because it allows any normal distribution to be converted into a common scale using z-scores, making it easier to calculate probabilities and compare different datasets.

13] What is the Central Limit Theorem (CLT), and why is it critical in statistics
- The Central Limit Theorem states that, regardless of a population's distribution, the distribution of the sample means approaches a normal distribution as the sample size increases, provided the samples are independent and identically distributed. It’s critical because it justifies using normal-based methods for inference even when the data isn't normally distributed

14] How does the Central Limit Theorem relate to the normal distribution
- The Central Limit Theorem explains why the normal distribution appears so often in statistics—it shows that the averages of large samples from any distribution tend to follow a normal distribution. This connection allows us to use normal-based methods for estimating probabilities and making inferences, even if the original data isn’t normally distributed.

15] What is the application of Z statistics in hypothesis testing
- Z statistics are used in hypothesis testing to determine how far a sample mean is from the population mean in terms of standard deviations, assuming the population variance is known. It's applied when testing large samples to decide whether to reject the null hypothesis based on the calculated z-score and its corresponding probability

16] How do you calculate a Z-score, and what does it represent
- A Z-score is calculated using the formula:
Z = (X − μ) / σ,
where X is the data point, μ is the mean, and σ is the standard deviation.

It represents how many standard deviations a value is from the mean, helping to understand if a value is typical or unusually high or low in a dataset.

17] What are point estimates and interval estimates in statistics
- Point estimates give a single best guess of a population parameter, like using a sample mean to estimate the population mean. Interval estimates provide a range of values, like a confidence interval, within which the true parameter is likely to lie, offering a measure of uncertainty around the estimate.

18] What is the significance of confidence intervals in statistical analysis
- Confidence intervals provide a range in which the true population parameter is likely to fall, based on sample data. They reflect the uncertainty of an estimate and help assess the reliability of conclusions, offering more insight than a single point estimate.

19] What is the relationship between a Z-score and a confidence interval
- A Z-score determines how many standard deviations a value is from the mean, and it is used to calculate confidence intervals when the population standard deviation is known. The Z-score corresponding to a confidence level (e.g., 1.96 for 95%) defines the range around the sample mean within which the true population mean is likely to lie

20] How are Z-scores used to compare different distributions
- Z-scores standardize values from different distributions by expressing them in terms of standard deviations from their respective means. This allows direct comparison of data points from different scales or units, helping identify which value is more extreme relative to its own distribution

21] What are the assumptions for applying the Central Limit Theorem
- The Central Limit Theorem assumes that samples are independent, identically distributed, and drawn randomly. Also, the sample size should be sufficiently large—typically 30 or more—for the sampling distribution of the mean to approximate a normal distribution, especially if the original population is not normal.

22] What is the concept of expected value in a probability distribution
- The expected value is the long-run average outcome of a random variable over many trials. It’s calculated by multiplying each possible value by its probability and summing the results. It represents the central tendency of a probability distribution

23] How does a probability distribution relate to the expected outcome of a random variable
- A probability distribution assigns probabilities to each possible value of a random variable, and the expected outcome is the weighted average of those values using the assigned probabilities. It tells you what result you would expect on average if the random process were repeated many times.

```python
# 3] Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): 0 or 1
    p (float): Probability of success (0 ≤ p ≤ 1)

    Returns:
    float: Probability of observing x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1")

    return p if x == 1 else 1 - p

print(bernoulli_pdf(1, 0.7))
```

```python
# 9] Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    """
    Calculate Z-scores and plot them.

    Parameters:
    data (list or array): Numeric dataset
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    z_scores = (data - mean) / std

    # Plotting
    plt.plot(z_scores, marker='o', linestyle='-', color='orange')
    plt.axhline(0, color='blue', linestyle='--')  # mean line
    plt.title("Z-scores of Dataset")
    plt.xlabel("Data Index")
    plt.ylabel("Z-score")
    plt.grid(True)
    plt.show()

    return z_scores

data = [10, 12, 9, 14, 13, 11, 15, 16, 10, 13]
z = calculate_and_plot_zscores(data)
print("Z-scores:", z)
```

```python
# 18] Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    """
    Calculate the Z-score of a data point.

    Parameters:
    x (float): Data point
    mean (float): Mean of the distribution
    std (float): Standard deviation of the distribution

    Returns:
    float: Z-score
    """
    return (x - mean) / std

# Example data
x_value = 72
mean = 70
std = 5

# Calculate Z-score
z = calculate_z_score(x_value, mean, std)
print(f"Z-score: {z:.2f}")

# Plot the standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.plot(x, y, label='Standard Normal PDF')
plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
plt.fill_between(x, 0, y, where=(x <= z), color='red', alpha=0.3)
plt.title('Z-score Position in Standard Normal Distribution')
plt.xlabel('Z')
plt.ylabel('Density')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 19] Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a one-sample Z-test.

    Parameters:
    sample_data (list or array): sample observations
    population_mean (float): hypothesized population mean
    population_std (float): known population standard deviation
    alpha (float): significance level

    Returns:
    None (prints result)
    """
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z:.2f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")

# Example usage
data = [68, 70, 72, 71, 69, 73, 75, 67, 70, 74]
pop_mean = 70
pop_std = 4  # Known population std

z_test(data, pop_mean, pop_std)
```

```python
# 20] Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate confidence interval for the mean of a dataset.

    Parameters:
    data (list or array): Sample data
    confidence (float): Confidence level (e.g., 0.95 for 95%)

    Returns:
    tuple: (lower_bound, upper_bound)
    """
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error
    margin = stats.t.ppf((1 + confidence) / 2, n - 1) * std_err

    lower = mean - margin
    upper = mean + margin

    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    return (lower, upper)

# Example usage
sample_data = [68, 70, 72, 71, 69, 73, 75, 67, 70, 74]
confidence_interval(sample_data, confidence=0.95)
```

```python
# 21] Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
from scipy import stats

# Step 1: Generate data from a normal distribution
np.random.seed(0)  # for reproducibility
mu = 100       # true mean
sigma = 15     # true standard deviation
n = 50         # sample size

data = np.random.normal(mu, sigma, n)

# Step 2: Calculate 95% confidence interval for the sample mean
sample_mean = np.mean(data)
std_err = stats.sem(data)
confidence = 0.95
margin = stats.t.ppf((1 + confidence) / 2, n - 1) * std_err
lower = sample_mean - margin
upper = sample_mean + margin

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
# 28] Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution
import numpy as np
from scipy.stats import skew, kurtosis

def compute_skewness_kurtosis(mean=0, std=1, size=1000):
    """
    Generate data from a normal distribution and compute skewness and kurtosis.

    Parameters:
    mean (float): Mean of the normal distribution
    std (float): Standard deviation
    size (int): Number of samples

    Returns:
    tuple: (skewness, kurtosis)
    """
    # Generate normal data
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness and kurtosis
    skew_val = skew(data)
    kurt_val = kurtosis(data)  # excess kurtosis (normal = 0)

    print(f"Skewness: {skew_val:.4f}")
    print(f"Kurtosis (excess): {kurt_val:.4f}")

    return skew_val, kurt_val

# Example usage
compute_skewness_kurtosis()
```
